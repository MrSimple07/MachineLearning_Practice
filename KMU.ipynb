```python
def longest_common_subsequence_length(str1, str2):
    m, n = len(str1), len(str2)

    # Create a table to store the lengths of LCS for subproblems
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    # Fill the table in a bottom-up manner
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if str1[i - 1] == str2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    return dp[m][n]

str1 = "HRTQYQDNJFKQHWCASFXGJRXPOMB"
str2 = "LSOPLAJTZSQYZBEZY"
lcs_length = longest_common_subsequence_length(str1, str2)

print(f"The length of the Longest Common Subsequence is: {lcs_length}")
```
