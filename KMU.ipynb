```python
from datasets import load_dataset

dataset = load_dataset("RussianNLP/tape", 'winograd.raw')
```

```python
dataset['train'][1]
```

```python
train_data = dataset['train'][:25]
train_questions = train_data['text']
train_answers = train_data['answer']

test_data = dataset['test'][:10]
test_questions = test_data['text']
test_answers = test_data['answer']

print("Training Set Questions and Answers:")
for i, (question, answer) in enumerate(zip(train_questions, train_answers)):
    print(f"{i + 1}. Question: {question}\n   Answer: {answer}")

print("\nTest Set Questions and Answers:")
for i, (question, answer) in enumerate(zip(test_questions, test_answers)):
    print(f"{i + 1}. Question: {question}\n   Answer: {answer}")
```

```python
from datasets import load_dataset

dataset = load_dataset('ai-forever/MERA', 'lcs')
```
